```python
import re
def cleanresume(txt):
  cleantxt= re.sub('http\s+\s',' ',txt)
  cleantxt= re.sub('@\s+',' ',cleantxt)
  cleantxt= re.sub('#http\s+\s',' ',txt)


  return cleantxt
```

```python
def clean_special_chars(txt):
    # Remove characters like *, ,, ., -, :, &, etc.
    return re.sub(r'[^\w\s,]', '', txt)
```

```python
import nltk
from nltk.corpus import stopwords

# Download the stopwords list from NLTK (run once)
nltk.download('stopwords')

# Define stop_words
stop_words = set(stopwords.words('english'))

```

```python
def remove_stopwords(txt):
    # Tokenize and remove stopwords
    return ' '.join([word for word in txt.split() if word.lower() not in stop_words])
```

```python
df['Cleanedresume'] = df['Resume'].apply(cleanresume)
df['Cleanedspecial'] = df['Cleanedresume'].apply(clean_special_chars)
df['updatedresume'] = df['Cleanedspecial'].apply(remove_stopwords)
df['updatedresume'][0]
```

WORDS INTO CATEGORICAL VALUES

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming the 'Category' column is already encoded
le = LabelEncoder()

# Fit the LabelEncoder and transform the categories
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])

# Create a mapping between numeric labels and original categories
category_mapping = pd.DataFrame({
    'Numeric_Label': range(len(le.classes_)),  # Numeric labels from 0 to n-1
    'Original_Category': le.classes_           # The corresponding original category names
})

# Save this mapping to a new CSV file
category_mapping.to_csv('category_mapping.csv', index=False)

# Display the mapping
print(category_mapping)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
tfidf.fit(df['updatedresume'])
resume=tfidf.transform(df['updatedresume'])
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'resume' and 'df['Category']' are defined
x_train, x_test, y_train, y_test = train_test_split(
    resume,
    df['Category'],
    test_size=0.33,
    random_state=42  # Correct argument name
)
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
clf= OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
print(ypred)
```

```python
print(accuracy_score(y_test,ypred))
```
